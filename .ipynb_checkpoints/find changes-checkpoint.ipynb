```python
import cv2
import matplotlib.pyplot as plt
```

```python
pre=cv2.imread("pre.jpg")
post=cv2.imread("post.jpg")
image= cv2.absdiff(pre, post)
```

```python
change_gray=cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(change_gray)
print(change_gray)
ret,thresh1 = cv2.threshold(change_gray,60,255,cv2.THRESH_BINARY)
```
